```python
# Cleaning and changing type of price and percentage columns from string to float64

# Prices with "," as thousands separator
marti['discounted_price'] = marti['discounted_price'].str.replace("₹",'')
marti['discounted_price'] = marti['discounted_price'].str.replace(",",'')
marti['discounted_price'] = marti['discounted_price'].astype('float64')

marti['actual_price'] = marti['actual_price'].str.replace("₹",'')
marti['actual_price'] = marti['actual_price'].str.replace(",",'')
marti['actual_price'] = marti['actual_price'].astype('float64')

# Converting indian rupee to US dollar
marti['actual_price'] = marti['actual_price']*0.012
marti['discounted_price'] = marti['discounted_price']*0.012

# Percentage from "PP%"" to PP/100 value
marti['discount_percentage'] = marti['discount_percentage'].str.replace('%','').astype('float64')
marti['discount_percentage'] = marti['discount_percentage'] / 100

# Rating Counts with "," as thousand separator
marti['rating_count'] = marti['rating_count'].str.replace(",",'').astype('float64')
```
